```python
import sys

for line in sys.stdin:
    a = list(map(int, line.split()))
    terms = []

    for i, coef in enumerate(a):
        deg = 8 - i

        if coef == 0:
            continue

        sign = "+" if coef > 0 else "-"
        val = abs(coef)

        # Build term
        if deg > 1:
            if val == 1:
                term = f"x^{deg}"
            else:
                term = f"{val}x^{deg}"
        elif deg == 1:
            if val == 1:
                term = "x"
            else:
                term = f"{val}x"
        else:
            term = str(val)

        # First term formatting
        if not terms:
            if coef < 0:
                terms.append("-" + term)
            else:
                terms.append(term)
        else:
            terms.append(f" {sign} {term}")

    # If all coefficients are zero
    if not terms:
        print("0")
    else:
        print("".join(terms))
```
