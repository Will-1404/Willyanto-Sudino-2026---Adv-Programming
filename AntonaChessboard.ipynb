```python
import math

while True:
    n = int(input())
    if n == 0:
        break

    k = math.ceil(math.sqrt(n))
    mid = k * k - k + 1

    if k % 2 == 0:
        if n >= mid:
            x = k
            y = k * k - n + 1
        else:
            x = n - (k - 1) * (k - 1)
            y = k
    else:
        if n >= mid:
            x = k * k - n + 1
            y = k
        else:
            x = k
            y = n - (k - 1) * (k - 1)

    print(x, y)
```
