```python
import sys

def solve():
    # Read from standard input
    for line in sys.stdin:
        # Parse the input line into integers a and b
        try:
            a, b = map(int, line.split())
        except ValueError:
            continue

        # Check for the termination condition
        if a == -1 and b == -1:
            break

        # Calculate the direct distance
        diff = abs(a - b)

        # Calculate the distance going the other way around the circle
        wrap_around_diff = 100 - diff

        # The result is the minimum of the two paths
        print(min(diff, wrap_around_diff))

if __name__ == "__main__":
    solve()
```
