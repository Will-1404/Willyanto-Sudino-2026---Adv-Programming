```python
vowels = "aeiouy"

while True:
    s = input()

    if s == "e/o/i":
        break

    parts = s.split('/')

    def count_syllables(line):
        count = 0
        inside = False

        for ch in line:
            if ch in vowels:
                if not inside:
                    count += 1
                    inside = True
            else:
                inside = False

        return count

    syllables = [count_syllables(part) for part in parts]

    if syllables[0] != 5:
        print(1)
    elif syllables[1] != 7:
        print(2)
    elif syllables[2] != 5:
        print(3)
    else:
        print("Y")
```
