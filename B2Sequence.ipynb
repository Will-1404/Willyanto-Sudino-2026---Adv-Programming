```python
#include <stdio.h>

int main() {
    int n, caseNum = 1;

    while (scanf("%d", &n) == 1) {
        int arr[105];

        for (int i = 0; i < n; i++) {
            scanf("%d", &arr[i]);
        }

        int isB2 = 1;

        // شرط 1: semua positif & strictly increasing
        if (arr[0] <= 0) isB2 = 0;

        for (int i = 1; i < n && isB2; i++) {
            if (arr[i] <= arr[i - 1]) {
                isB2 = 0;
            }
        }

        // شرط 2: semua pair sum unik
        int seen[20005] = {0}; // max sum = 10000 + 10000

        for (int i = 0; i < n && isB2; i++) {
            for (int j = i; j < n; j++) {
                int sum = arr[i] + arr[j];
                if (seen[sum]) {
                    isB2 = 0;
                    break;
                }
                seen[sum] = 1;
            }
        }

        // Output
        if (isB2) {
            printf("Case #%d: It is a B2-Sequence.\n\n", caseNum);
        } else {
            printf("Case #%d: It is not a B2-Sequence.\n\n", caseNum);
        }

        caseNum++;
    }

    return 0;
}
```
