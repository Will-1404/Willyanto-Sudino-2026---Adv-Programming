```python
case = 1

while True:
    s = input().strip()

    if s == "#":
        break

    if s == "HELLO":
        lang = "ENGLISH"
    elif s == "HOLA":
        lang = "SPANISH"
    elif s == "HALLO":
        lang = "GERMAN"
    elif s == "BONJOUR":
        lang = "FRENCH"
    elif s == "CIAO":
        lang = "ITALIAN"
    elif s == "ZDRAVSTVUJTE":
        lang = "RUSSIAN"
    else:
        lang = "UNKNOWN"

    print(f"Case {case}: {lang}")
    case += 1
```
