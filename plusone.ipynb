```python
def plusOne(digits):
    # start from the last digit
    for i in range(len(digits) - 1, -1, -1):
        if digits[i] < 9:
            digits[i] += 1
            return digits
        else:
            digits[i] = 0

    # if all digits were 9
    return [1] + digits


# Test the examples
examples = [
    [1,2,3],
    [4,3,2,1],
    [9]
]

for digits in examples:
    result = plusOne(digits.copy())
    print("Input:", digits)
    print("Output:", result)
    print()
```
