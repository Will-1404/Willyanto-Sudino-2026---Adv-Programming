```python
from collections import deque

# Directions: N, E, S, W
dirs = [(-1, 0, 'N'), (0, 1, 'E'), (1, 0, 'S'), (0, -1, 'W')]

while True:
    sx, sy = map(int, input().split())

    if sx == 0 and sy == 0:
        break

    ex, ey = map(int, input().split())

    # Convert to 0-based indexing
    sx -= 1
    sy -= 1
    ex -= 1
    ey -= 1

    # blocked[r][c][d]
    # d = 0:N, 1:E, 2:S, 3:W
    blocked = [[[False] * 4 for _ in range(6)] for _ in range(6)]

    for _ in range(3):
        x1, y1, x2, y2 = map(int, input().split())

        # Vertical wall
        if x1 == x2:
            col = x1
            for row in range(min(y1, y2), max(y1, y2)):
                r = row
                c1 = col - 1
                c2 = col

                if 0 <= r < 6 and 0 <= c1 < 6:
                    blocked[r][c1][1] = True  # East blocked
                if 0 <= r < 6 and 0 <= c2 < 6:
                    blocked[r][c2][3] = True  # West blocked

        # Horizontal wall
        else:
            row = y1
            for col in range(min(x1, x2), max(x1, x2)):
                r1 = row - 1
                r2 = row
                c = col

                if 0 <= r1 < 6 and 0 <= c < 6:
                    blocked[r1][c][2] = True  # South blocked
                if 0 <= r2 < 6 and 0 <= c < 6:
                    blocked[r2][c][0] = True  # North blocked

    # BFS
    q = deque()
    q.append((sy, sx))
    visited = [[False] * 6 for _ in range(6)]
    visited[sy][sx] = True

    parent = {}

    while q:
        r, c = q.popleft()

        if (r, c) == (ey, ex):
            break

        for i, (dr, dc, ch) in enumerate(dirs):
            nr, nc = r + dr, c + dc

            if 0 <= nr < 6 and 0 <= nc < 6:
                if not blocked[r][c][i] and not visited[nr][nc]:
                    visited[nr][nc] = True
                    parent[(nr, nc)] = (r, c, ch)
                    q.append((nr, nc))

    # Reconstruct path
    path = []
    cur = (ey, ex)

    while cur != (sy, sx):
        pr, pc, ch = parent[cur]
        path.append(ch)
        cur = (pr, pc)

    print(''.join(path[::-1]))
```
