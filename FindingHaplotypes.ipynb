```python
from itertools import product

def possible_pairs(genotype):
    idx = [i for i, c in enumerate(genotype) if c == '2']
    pairs = []

    for bits in product('01', repeat=len(idx)):
        h1 = list(genotype)
        h2 = list(genotype)

        for i, b in zip(idx, bits):
            h1[i] = b
            h2[i] = '1' if b == '0' else '0'

        for i in range(len(genotype)):
            if genotype[i] == '0':
                h1[i] = h2[i] = '0'
            elif genotype[i] == '1':
                h1[i] = h2[i] = '1'

        pairs.append(("".join(h1), "".join(h2)))

    return pairs


while True:
    try:
        N, L = map(int, input().split())

        if N == 0 and L == 0:
            break

        genotypes = [input().strip() for _ in range(N)]

        all_pairs = [possible_pairs(g) for g in genotypes]

        best = [float('inf')]  # use list instead of nonlocal

        def backtrack(i, used):
            if len(used) >= best[0]:
                return

            if i == N:
                best[0] = min(best[0], len(used))
                return

            for a, b in all_pairs[i]:
                new_used = used | {a, b}
                backtrack(i + 1, new_used)

        backtrack(0, set())

        print(best[0])

    except EOFError:
        break
```
