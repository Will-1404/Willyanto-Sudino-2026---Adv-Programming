```python
def is_happy(n):
    seen = set()

    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(digit) ** 2 for digit in str(n))

    return n == 1


t = int(input())

for i in range(1, t + 1):
    n = int(input())

    if is_happy(n):
        print(f"Case #{i}: {n} is a Happy number.")
    else:
        print(f"Case #{i}: {n} is an Unhappy number.")
```
