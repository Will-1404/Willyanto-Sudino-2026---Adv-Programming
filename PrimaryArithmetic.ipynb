```python
import sys

def count_carries(a, b):
    carry = 0
    count = 0

    # Process digits from right to left
    while a > 0 or b > 0:
        digit_sum = (a % 10) + (b % 10) + carry

        if digit_sum >= 10:
            carry = 1
            count += 1
        else:
            carry = 0

        a //= 10
        b //= 10

    return count


for line in sys.stdin:
    x, y = map(int, line.split())

    if x == 0 and y == 0:
        break

    carries = count_carries(x, y)

    if carries == 0:
        print("No carry operation.")
    elif carries == 1:
        print("1 carry operation.")
    else:
        print(f"{carries} carry operations.")
```
