```python
def isValid(s):
    stack = []
    mapping = {
        ')': '(',
        ']': '[',
        '}': '{'
    }

    for char in s:
        if char in mapping:  # closing bracket
            if not stack or stack[-1] != mapping[char]:
                return False
            stack.pop()
        else:  # opening bracket
            stack.append(char)

    return len(stack) == 0


# Let user input
while True:
    s = input("Enter brackets string (or type 'exit' to quit): ")

    if s.lower() == 'exit':
        break

    print(isValid(s))
```

```python
def isValid(s):
    while "()" in s or "[]" in s or "{}" in s:
        s = s.replace("()", "").replace("[]", "").replace("{}", "")
    return s == ""


# Let user input
while True:
    s = input("Enter brackets string (or type 'exit' to quit): ")

    if s.lower() == 'exit':
        break

    print(isValid(s))
```
