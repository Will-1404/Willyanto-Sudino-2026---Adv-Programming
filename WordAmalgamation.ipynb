```python
import sys

def solve():
    dictionary = {}

    # Process phase 1: Build the dictionary
    for line in sys.stdin:
        line = line.strip()
        if line == 'XXXXXX':
            break

        # Sort characters to create a unique key for anagrams
        sorted_word = "".join(sorted(line))

        # Add to dictionary mapping
        if sorted_word not in dictionary:
            dictionary[sorted_word] = []
        dictionary[sorted_word].append(line)

    # Sort the list of found words alphabetically for the required output
    for key in dictionary:
        dictionary[key].sort()

    # Process phase 2: Unscramble inputs
    for line in sys.stdin:
        line = line.strip()
        if line == 'XXXXXX':
            break

        sorted_input = "".join(sorted(line))

        if sorted_input in dictionary:
            # Print all anagrams found in dictionary
            for word in dictionary[sorted_input]:
                print(word)
        else:
            # No valid anagrams found
            print("NOT A VALID WORD")

        # Signal the end of the list
        print("******")

if __name__ == "__main__":
    solve()
```
