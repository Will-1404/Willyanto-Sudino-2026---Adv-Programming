```python
from collections import Counter
import sys

lines = sys.stdin.read().splitlines()

for i in range(0, len(lines), 2):
    s1 = lines[i]
    s2 = lines[i + 1]

    # Compare frequency patterns
    if sorted(Counter(s1).values()) == sorted(Counter(s2).values()):
        print("YES")
    else:
        print("NO")
```
