```python
MOD = 1000007

t = int(input())

for tc in range(1, t + 1):
    n = int(input())
    board = [input().strip() for _ in range(n)]

    # find white piece
    for r in range(n):
        for c in range(n):
            if board[r][c] == 'W':
                sr, sc = r, c

    dp = [[0] * n for _ in range(n)]
    dp[sr][sc] = 1

    # move upward (towards row 0)
    for r in range(sr, 0, -1):
        for c in range(n):
            if dp[r][c] == 0:
                continue

            # up-left
            nr, nc = r - 1, c - 1
            if 0 <= nr < n and 0 <= nc < n:
                if board[nr][nc] == '.':
                    dp[nr][nc] = (dp[nr][nc] + dp[r][c]) % MOD
                elif board[nr][nc] == 'B':
                    jr, jc = r - 2, c - 2
                    if 0 <= jr < n and 0 <= jc < n and board[jr][jc] == '.':
                        dp[jr][jc] = (dp[jr][jc] + dp[r][c]) % MOD

            # up-right
            nr, nc = r - 1, c + 1
            if 0 <= nr < n and 0 <= nc < n:
                if board[nr][nc] == '.':
                    dp[nr][nc] = (dp[nr][nc] + dp[r][c]) % MOD
                elif board[nr][nc] == 'B':
                    jr, jc = r - 2, c + 2
                    if 0 <= jr < n and 0 <= jc < n and board[jr][jc] == '.':
                        dp[jr][jc] = (dp[jr][jc] + dp[r][c]) % MOD

    ans = sum(dp[0]) % MOD
    print(f"Case {tc}: {ans}")
```
