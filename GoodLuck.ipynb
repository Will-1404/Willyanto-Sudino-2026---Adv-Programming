```python
t = int(input())

for case in range(1, t + 1):
    sites = []
    max_val = 0

    for _ in range(10):
        url, v = input().split()
        v = int(v)
        sites.append((url, v))
        max_val = max(max_val, v)

    print(f"Case #{case}:")
    for url, v in sites:
        if v == max_val:
            print(url)
```
